```
from bs4 import BeautifulSoup

text_scrap = ''
with open('data.html','r') as html_file: #r read, w write
  content = html_file.read() #membaca semua file

  soup = BeautifulSoup(content, 'lxml')
  course_html_tags = soup.find_all('div', class_ = 'ants-news-section ants-news-content min-height:300px;')
  for course in course_html_tags:
    text_scrap += str(course) + '\n'
    
print(text_scrap)
```

**Preprocessing**

```
!pip install nltk==3.5
```

Case Folding

```
Lower_string = text_scrap.lower()
print(Lower_string)
```

Tokenizing

```
import nltk
nltk.download('punkt')
import string 
import re #regex library
from nltk.tokenize import word_tokenize 
```

```
#remove angka
Lower_string = re.sub(r"\d+", "", Lower_string)
print(Lower_string)
```

```
#remove punctuation
Lower_string = Lower_string.translate(str.maketrans("","",string.punctuation))
print(Lower_string)
```

```
#remove whitespace leading & trailing
Lower_string = Lower_string.strip()
print(Lower_string)
```

```
#remove multiple whitespace into single whitespace
Lower_string = re.sub('\s+',' ',Lower_string)
print(Lower_string)
```

```
tokens = nltk.tokenize.word_tokenize(Lower_string)
print('Tokenizing Result : \n') 
print(tokens)
```

```
# import word_tokenize & FreqDist from NLTK
from nltk.tokenize import word_tokenize 
from nltk.probability import FreqDist
```

```
freq_tokens = nltk.FreqDist(tokens)

print('Frequency Tokens : \n') 
print(freq_tokens.most_common())
```

```
import pandas as pd

df_freq_tokens = pd.DataFrame.from_dict(freq_tokens, orient='index')
df_freq_tokens.columns = ['Frequency']
df_freq_tokens.index.name = 'Key'

df_freq_tokens.plot(kind='bar')
```

Filtering

```
nltk.download('stopwords')
```

```
from nltk.corpus import stopwords

# tokenize text
freq_tokens

list_stopwords = set(stopwords.words('indonesian'))

#remove stopword pada list token
tokens_without_stopword = [word for word in freq_tokens if not word in list_stopwords]


print(tokens_without_stopword)
```

Stemming

```
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```
stemmer = PorterStemmer()
```

```
# token without stopword
list_tokens = tokens_without_stopword

# stem
output   = [(token + " : " + stemmer.stem(token)) for token in list_tokens]

output
```

Mencari index

```
index_tag=[]

for i in range(0, len(output)):
  if output[i].find('ants') != -1:
    index_tag.append(i)
  elif output[i].find('news') != -1:
    index_tag.append(i)
  elif output[i].find('div') != -1:
    index_tag.append(i) 
  elif output[i].find('class') != -1:
    index_tag.append(i)
  elif output[i].find('strongdot') != -1:
    index_tag.append(i)
  elif output[i].find('minheight') != -1:
    index_tag.append(i)
  elif output[i].find('style') != -1:
    index_tag.append(i)
  elif output[i].find('margin') != -1:
    index_tag.append(i)

print(index_tag)
```

Menghapus elemen

```
j = 0

for index in index_tag:
  id = index-j
  del output[id]
  j+=1

print(output)
```
